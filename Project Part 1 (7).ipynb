```python
# 篩選特定日期的數據
selected_date = "2019-07-14"
filtered_treatment_data = treatments_data_selected.loc[
    (treatments_data_selected['date_time'] >= pd.to_datetime(selected_date)) &
    (treatments_data_selected['date_time'] < pd.to_datetime(selected_date) + pd.Timedelta(days=1))
]

print(filtered_treatment_data.to_string())
```

# Filter information from 2019-07-14

```python
selected_date = "2019-07-14"
filtered_sgv_data = sgv_data.loc[
    (sgv_data['date_time'] >= pd.to_datetime(selected_date)) &
    (sgv_data['date_time'] < pd.to_datetime(selected_date) + pd.Timedelta(days=1))
]

print(filtered_sgv_data.to_string())
```

```python
# outer join
merged_data = pd.merge(
    filtered_sgv_data,
    filtered_treatment_data,
    on='date_time',
    how='outer' 
)

# 按時間排序
merged_data = merged_data.sort_values(by='date_time')

merged_data
```

```python
print(merged_data['date_time'].isna().sum())
```

```python
print(type(merged_data))
print(merged_data.columns)
print(merged_data.dtypes)
```

```python
print(merged_data.index.duplicated().sum()) 
print(merged_data[merged_data.index.duplicated(keep=False)])
```

```python
# 1. Set datetime as index
#merged_data['date_time'] = pd.to_datetime(merged_data['date_time'])
#merged_data.set_index('date_time', inplace=True)

# 2. Sample blood glucose data every five minutes
sgv_resampled = merged_data['sgv'].resample('5T').mean().reset_index()

# 3. Leave the rest of the data as is
carbs_insulin = merged_data[['carbs', 'insulin']].reset_index()

# 4. Merge resampled Glucose and original Carbs/Insulin data
# Use merge_asof for approximate matching so that Carbs and Insulin maintain the original resolution
result = pd.concat([sgv_resampled, carbs_insulin]).sort_values(by='date_time').reset_index(drop=True)

# 5. If necessary, sort the results by time (in case the order changes)
result.sort_values(by='date_time', inplace=True)

# 6. fill in the missing values ​​​​of sgv
result['sgv'] = result['sgv'].interpolate(method='linear')

result
```

# normalized data

```python
for col in ['sgv', 'carbs', 'insulin']:
    result[f'{col}_normalized'] = (result[col] - result[col].min()) / (result[col].max() - result[col].min())
```

# Draw charts

```python
# 確保數據中沒有 NaN
result[['sgv_normalized', 'insulin_normalized', 'carbs_normalized']] = result[['sgv_normalized', 'insulin_normalized', 'carbs_normalized']].interpolate()

# 繪製圖表，強制加入連線
plt.figure(figsize=(12, 6))

plt.plot(result['date_time'], result['sgv_normalized'], label='Blood Sugar (mg/dL) - Normalized', color='blue', marker='o', linestyle='-')
plt.plot(result['date_time'], result['insulin_normalized'], label='Bolus Insulin Dose - Normalized', color='green', marker='s', linestyle='-')
plt.plot(result['date_time'], result['carbs_normalized'], label='Carbohydrate Intake (g) - Normalized', color='orange', marker='^', linestyle='-')

plt.title('Blood Sugar, Insulin, and Carbohydrate Intake Over Time')
plt.xlabel('Time (HH:MM:SS)')
plt.ylabel('Normalized Value')
plt.legend()
plt.grid()
plt.tight_layout()

plt.show()
```
